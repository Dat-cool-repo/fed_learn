```python
import os
from dotenv import load_dotenv

# Define the path to your .env file in Google Drive
dotenv_path = '/content/drive/MyDrive/fed_learn/.env'

# Load environment variables from the specified .env file
load_dotenv(dotenv_path=dotenv_path)

# Retrieve the HF_TOKEN from environment variables
hf_token = os.getenv('HF_TOKEN')

# Set it as an environment variable (redundant if already loaded, but good for clarity)
os.environ['HF_TOKEN'] = hf_token

if hf_token:
    print("Hugging Face token has been loaded from .env and set successfully!")
else:
    print(f"Warning: HF_TOKEN not found in .env file at {dotenv_path} or environment variables.")
```
